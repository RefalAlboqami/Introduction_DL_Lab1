# Step 3: Data Cleaning and Preprocessing

```python
print('\nNull:\n',df.isnull().sum())
print('\nduplicated:', df.duplicated().sum())
```

```python
df.drop(columns=['customer name', 'customer e-mail','gender'],axis=1,inplace=True)
df
```

```python
df['age'] = df['age'].astype(int)
print(df.dtypes)
```

```python
cat_col=['country']

df = pd.get_dummies(df, columns=cat_col, dtype=int)
df
```

```python
numeric = ['age', 'annual Salary', 'credit card debt', 'net worth', 'car purchase amount']

scaler = StandardScaler()
df[numeric] = scaler.fit_transform(df[numeric])
df
```

```python
scaler = MinMaxScaler()
df[numeric] = scaler.fit_transform(df[numeric])
df
```

# Step 4: Train Test Split

```python
features = ['annual Salary', 'credit card debt', 'net worth']

plt.figure(figsize=(8, 6))
df[features].boxplot()
plt.title('Boxplot')
plt.xlabel('Data')
plt.show()
```

```python
def limits(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1

    upper = Q3 + 1.5*IQR
    lower = Q1 - 1.5*IQR

    print("Upper Bound:",upper)
    print("Lower Bound:",lower)
    return upper, lower

print("annual Salary")
upper1, lower1 = limits(df['annual Salary'])
print("----------------------------------")
print("credit card debt")
upper2, lower2 = limits(df['credit card debt'])
print("----------------------------------")
print("net worth")
upper3, lower3 = limits(df['net worth'])
```

```python
outliers = df[(df['annual Salary'] < lower1) | (df['annual Salary'] > upper1)]
df = df.drop(outliers.index)

outliers = df[(df['credit card debt'] < lower2) | (df['credit card debt'] > upper2)]
df = df.drop(outliers.index)

outliers = df[(df['net worth'] < lower3) | (df['net worth'] > upper3)]
df = df.drop(outliers.index)
```

```python
plt.figure(figsize=(8, 6))
df[features].boxplot()
plt.title('Boxplot')
plt.xlabel('Data')
plt.show()
```

```python
X = df.drop(['car purchase amount'], axis=1)
y = df['car purchase amount']

X_train, X_test, y_train, y_test = train_test_split(X,y , random_state=40,test_size=0.20, shuffle=True)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Step 5: Build the Artifical Neural Network Model

```python
model = Sequential()
model.add(Dense(64, input_shape=(215,)))
model.add(Dense(units=32, activation='relu', input_shape=(3,)))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(
              optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

# Step 6: Train the Model


```python
X_train_tf = tf.constant(X_train, dtype=tf.float32)
y_train_tf = tf.constant(y_train, dtype=tf.float32) 
model.fit(X_train_tf, y_train_tf, epochs=10, batch_size=32, validation_split=0.2)
```

# Step 7: Evaluate the Model

```python
X_test_tf = tf.constant(X_test, dtype=tf.float32)
y_test_tf = tf.constant(y_test, dtype=tf.float32) 

evaluate1 = model.evaluate(X_test_tf, y_test_tf)
evaluate1
```

# Step 8: Build the Deep Neural Network Model

```python
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(
              optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```
